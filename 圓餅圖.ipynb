```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
values = [1330,550,350,400,250]
labels = ["Home mortgage payment","Food","Phone/Internet","Car","Other"]
explode = [0.1,0.1,0,0,0] #凸顯某區塊
```

```python
plt.pie(values,labels = labels,radius = 1.5,textprops = {"fontsize":14},autopct = "%0.1f%%",explode = explode,shadow = True,startangle = 100) #radius 可以調整半徑  textprops:以字典形式存在   autopct:顯示百分比 型態為str
#顯示到小數點後一位"%0.1f%%" startangle:從幾度開始顯示
#plt.show()
plt.savefig("pie.png",bbox_inches = "tight",pad_inches = 1,transparent = True) #bbox_inches 設定大小 #pad_inches = 添加外邊距 #
```
