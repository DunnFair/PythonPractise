```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
company = ["google","Amazon","Micrisoft","Facebook"]
revenue = [90,136,89,27]
profit = [40,2,34,12]
```

```python
xpos = np.arange(len(company))
```

```python
xpos
```

```python
plt.bar(xpos,revenue)
```

```python
plt.xticks(xpos,company) #xticks:x軸所在的刻度
plt.xlabel("Revenus(bin)")
plt.title("US tech Stocks")


plt.bar(xpos,revenue,label = "Revenus")
plt.bar(xpos,profit,label = "Profit")
plt.legend()
```

```python
#把柱狀體分開

plt.xticks(xpos,company) #xticks:x軸所在的刻度
plt.xlabel("Revenus(bin)")
plt.title("US tech Stocks")


plt.bar(xpos-0.2,revenue,width = 0.4,label = "Revenus")
plt.bar(xpos+0.2,profit,width = 0.4,label = "Profit")
plt.legend()
```
