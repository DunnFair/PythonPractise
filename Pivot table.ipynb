```python
import pandas as pd
import numpy as np
df = pd.read_csv(r"C:\Users\Darren\Desktop\PythonPratic\City+Weather.csv")
df
```

```python
#取出上述表中每一個城市，他們不同的天氣狀況(希望以時間作為索引，以城市作為主要的觀看內容)
df.pivot(index = "Date",columns = "City")
```

```python
#只需要其中一個欄位資料時 例如只要顯示溫度
df.pivot(index = "Date",columns = "City",values = "Temperature")
```

```python
#作其中數據做為加法運算或是取平均值(看似與上述類似依樣，但是這邊我們後面可以加一些總結的來未)
df.pivot_table(index = "Date",columns = "City",values = "Temperature",margins = True)
#例如要找7/30號這三個城市的溫度平均值(右方的ALL 試算出平均值 下方也是)
```

```python
#例如要得到總和的話後面加入SUM
df.pivot_table(index = "Date",columns = "City",values = "Temperature",margins = True,aggfunc = "sum")
```

```python
#例如要得到總和的話後面加入SUM(使用numpy方法)
df.pivot_table(index = "Date",columns = "City",values = "Temperature",margins = True,aggfunc = np.mean)
```

```python
#整理出七月份溫度的一些變化以及八月分各個城市的變化，依照每個月時間總結
#確認Date是否為時間格式
df["Date"] = pd.to_datetime(df.Date)
```

```python
df
```

```python
#依照時間做分組 Freq = "M"為一個月 定義一個key 需要用哪一個Column作為 算出平均值
df.pivot_table(index = pd.Grouper(freq="M",key = "Date"),columns = "City",values = "Temperature")
```

```python
#依照時間做分組 Freq = "M"為一個月 定義一個key 需要用哪一個Column作為 算出最大值
df.pivot_table(index = pd.Grouper(freq="M",key = "Date"),columns = "City",values = "Temperature",aggfunc = np.max)
```
