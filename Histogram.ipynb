```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blood_sugar_men = [113,85,90,150,149,88,93,115,135,80,77,82,129]
blood_sugar_women = [67,98,89,120,133,150,84,69,89,75,120,112,100]
```

```python
plt.xlabel("sugar range")
plt.ylabel("Total no of patients")
plt.title("Blood Sugar analysis")
plt.hist([blood_sugar_men,blood_sugar_women],bins = [80,100,125,150],rwidth = 0.95,color = ["r","g"],label = ["Men","Women"]) #第一個array是 Y軸 第二個是X軸
plt.legend()
```
