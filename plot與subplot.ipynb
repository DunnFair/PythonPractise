```python
import matplotlib.pyplot as plt
```

```python
#使用matplotlib inline 因為這個可以讓他在Jupyter 顯示圖表並且保存
```

```python
%matplotlib inline
```

```python
import numpy as np 
```

```python
x = np.arange(25)
```

```python
y = x ** 2 #平方
```

```python
x
```

```python
y
```

```python
plt.plot(x,y,"r--") #(x軸,y軸,"顏色--") 加上"--"可以變虛線
plt.xlabel("x Label")
plt.ylabel("y Label")
plt.title("my first plot")
plt.show()
```

```python
plt.subplot(1,2,1) #(row,column,index) 這個空間內佔據多少位子 從左到右
plt.plot(x,y,"b")

plt.subplot(1,2,2)
plt.plot(y,x,"y")
```

```python
plt.subplot(2,2,1) #(row,column,index) 這個空間內佔據多少位子 從左到右
plt.plot(x,y,"b")

plt.subplot(2,2,2)
plt.plot(y,x,"y")

plt.subplot(2,2,3) #(row,column,index) 這個空間內佔據多少位子 從左到右
plt.plot(x,y,"b--")

plt.subplot(2,2,4)
plt.plot(y,x,"y--")
```
