```python
import pandas as pd
df = pd.read_csv(r"C:\Users\Darren\Desktop\PythonPratic\City+Weather.csv")
df
```

```python
#id_vars 想要在新的DF當中不變得Columns 例如 在新的數據表裡面有日期的Date
df1 = pd.melt(df,id_vars = ["Date","City"])

df1
#pd.melt?
```

```python
#如果只是要顯示某個部份的話可以新增條件，只顯示溫度的部分
df1 = pd.melt(df,id_vars = ["Date"])
df1[df1["variable"] == "Temperature"]
```

```python
#不只是Date不改變還想要增加城市也不改變的話,會有順序問題
df1 = pd.melt(df,id_vars = ["Date","City"],value_vars = ["Temperature"])
df1
```

```python
#如果我們部輸入value_vars的話 並且改動Column欄位名稱
df1 = pd.melt(df,id_vars = ["Date","City"],var_name = "Property",value_name = "value Name")
#pd.melt?
df1
```
